### Question 2

Write a function that fits a polynomial LinearRegression model on the training data `X_train` for degrees 0 through 9. For each model compute the $R^2$ (coefficient of determination) regression score on the training data as well as the the test data, and return both of these arrays in a tuple.

*This function should return one tuple of numpy arrays `(r2_train, r2_test)`. Both arrays should have shape `(10,)`*

```python
def answer_two():
    from sklearn.linear_model import LinearRegression
    from sklearn.preprocessing import PolynomialFeatures
    from sklearn.metrics.regression import r2_score

    # Your code here
    
    #Create 2D array
    X=x.reshape(-1,1)
    
    #Initate a list to store arrays of 10 R-squared
    train_scores=[]
    test_scores=[]
    r2_scores=[]
    #Create polynomial features
    for i, degree in enumerate(range(10)):
        poly = PolynomialFeatures(degree=degree)
        X_poly = poly.fit_transform(X)

        X_train, X_test, y_train, y_test = train_test_split(X_poly, y,
                                                           random_state = 0)
        #Fit linear regression model on polynomial transformed data
        linreg = LinearRegression().fit(X_train, y_train)

        #Get R-squared
        train_scores.append(linreg.score(X_train, y_train))
        test_scores.append(linreg.score(X_test, y_test))
        r2_scores.append(r2_score(y_test, linreg.predict(X_test))) #same as test_scores

    return (np.asarray(train_scores), np.asarray(test_scores))
```

### Question 4

Training models on high degree polynomial features can result in overly complex models that overfit, so we often use regularized versions of the model to constrain model complexity, as we saw with Ridge and Lasso linear regression.

For this question, train two models: a non-regularized LinearRegression model (default parameters) and a regularized Lasso Regression model (with parameters `alpha=0.01`, `max_iter=10000`) both on polynomial features of degree 12. Return the $R^2$ score for both the LinearRegression and Lasso model's test sets.

*This function should return one tuple `(LinearRegression_R2_test_score, Lasso_R2_test_score)`*

```python
def answer_four():
    from sklearn.preprocessing import PolynomialFeatures
    from sklearn.linear_model import Lasso, LinearRegression
    from sklearn.metrics.regression import r2_score

    # Your code here
    #Create 2D array
    X=x.reshape(-1,1)

    #Create polynomial features
    poly = PolynomialFeatures(degree=12)
    X_poly = poly.fit_transform(X)

    X_train_poly, X_test_poly, y_train, y_test = train_test_split(X_poly, y,
                                                                   random_state = 0)

    #Scale polynomial features
    from sklearn.preprocessing import MinMaxScaler, StandardScaler
#     scaler = MinMaxScaler()
    scaler = StandardScaler()
    X_train_poly_scaled = scaler.fit_transform(X_train_poly)
    X_test_poly_scaled = scaler.transform(X_test_poly)

    #Fit linear regression model on polynomial transformed data
    linreg = LinearRegression().fit(X_train_poly, y_train)
    linlasso = Lasso(alpha=0.01, max_iter = 10000).fit(X_train_poly, y_train)
    linlasso_scaled= Lasso(alpha=0.01, max_iter = 10000).fit(X_train_poly_scaled, y_train)

    #Get R-squared
    linreg_score = r2_score(y_test, linreg.predict(X_test_poly))
    linlasso_score = r2_score(y_test, linlasso.predict(X_test_poly))
    linlasso_scaled_score = r2_score(y_test, linlasso_scaled.predict(X_test_poly_scaled))
    
    #Plot predictions
    lr_train = pd.DataFrame(list(zip(list(map(lambda k: k[1], X_train_poly)), linreg.predict(X_train_poly))),
            columns=['x', 'y_pred']).sort_values(by='x')
    lr_test = pd.DataFrame(list(zip(list(map(lambda k: k[1], X_test_poly)), linreg.predict(X_test_poly))),
            columns=['x', 'y_pred']).sort_values(by='x')

    
    df_train = pd.DataFrame(list(zip(list(map(lambda k: k[1], X_train_poly)), linlasso.predict(X_train_poly))),
            columns=['x', 'y_pred']).sort_values(by='x')
    df_test = pd.DataFrame(list(zip(list(map(lambda k: k[1], X_test_poly)), linlasso.predict(X_test_poly))),
            columns=['x', 'y_pred']).sort_values(by='x')

    def plot_prediction():
        import matplotlib.pyplot as plt
        %matplotlib notebook
        plt.figure(figsize=(10,5))
        plt.plot(X_train, y_train, 'o', label='train data')
        plt.plot(X_test, y_test, 'o', label='test data')

        plt.plot(lr_train['x'], lr_train['y_pred'], '-', color='Grey', label='Linear Regression train data')
        plt.plot(lr_test['x'], lr_test['y_pred'], '-', color='tab:Pink', label='Linear Regression test data')

        plt.plot(df_train['x'], df_train['y_pred'], '-', color='LightBlue', label='Lasso train data')
        plt.plot(df_test['x'], df_test['y_pred'], '-', color='tab:Orange', label='Lasso test data')
        plt.legend();
    #plot_prediction()

    return (linreg_score, linlasso_score) #, linlasso_scaled_score)

    
#answer_four()    
```
